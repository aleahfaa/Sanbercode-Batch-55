<h2>Soal 1: Pemahaman Component Matplotlib</h2>

- Jelaskan apa itu Figure
- Jelaskan apa itu Axis

Sebuah Figure dalam Matplotlib adalah wadah tingkat atas yang menyimpan semua elemen dari sebuah plot. Ini berfungsi sebagai kanvas di mana plot digambar dan mewakili seluruh jendela atau halaman di mana plot digambar. Anda dapat memikirkannya sebagai lembar kertas kosong di mana Anda membuat visualisasi Anda.

Sebuah Axis dalam Matplotlib adalah area berbentuk persegi panjang di dalam gambar di mana data diplot. Setiap gambar dapat berisi satu atau lebih axis, yang disusun dalam baris dan kolom jika diperlukan. Axis menyediakan sistem koordinat dan tempat sebagian besar plot dilakukan. Objek axis mewakili axis-x dan axis-y dari plot. Mereka menentukan batas data, lokasi tanda centang, label tanda centang, dan label axis. Setiap axis memiliki skala dan pengatur yang menentukan bagaimana tanda centang diatur.

<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# membuat Figure dan Axis dengan Data Kosong
fig=plt.figure()
ax=fig.add_subplot()
plt.show()
```

```python
#eksplisit
fig_explicit, ax_explicit=plt.subplots()
ax_explicit.plot(x, y)
plt.show()
```

```python
#implicit
plt.plot(x, y)
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
import seaborn as sns

sns.set()
plt.plot(x, y, label='cos(x/2)')
plt.plot(x, y2, label='sin(x)')
plt.plot(x, y3, label='sin(2x)')
plt.legend()

plt.show()
```
