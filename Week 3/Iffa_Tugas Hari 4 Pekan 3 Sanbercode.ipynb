<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk Global_Sales berdasarkan kategori Genre.

<h2>Soal 2: Perbandingan Kuantitatif Pieplot - Part a whole using Pie plot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here

plt.figure(figsize=(10, 8))
plt.pie(genre_globalsales['Global_Sales'], labels=genre_globalsales['Genre'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
```

<h2>Soal 3: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
genre_noglobal=df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean().reset_index()
genre_noglobal
```

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
plt.figure(figsize=(12, 8))

genres = genre_noglobal['Genre']
na_sales = genre_noglobal['NA_Sales']
eu_sales = genre_noglobal['EU_Sales']
jp_sales = genre_noglobal['JP_Sales']
other_sales = genre_noglobal['Other_Sales']

x = range(len(genres))

plt.bar(x, na_sales, width=0.2, align='center', label='NA_Sales')
plt.bar([i + 0.2 for i in x], eu_sales, width=0.2, align='center', label='EU_Sales')
plt.bar([i + 0.4 for i in x], jp_sales, width=0.2, align='center', label='JP_Sales')
plt.bar([i + 0.6 for i in x], other_sales, width=0.2, align='center', label='Other_Sales')

plt.xlabel('Genre')
plt.ylabel('Average Sales')
plt.title('Average Sales by Genre and Region')
plt.xticks([i + 0.3 for i in x], genres, rotation=90)
plt.legend()

plt.tight_layout()
plt.show()
```

Saya dapat mengetahui dengan mudah melalui visualisasi data rata-rata penjualan video game berdasarkan genre untuk setiap region kecuali Global_Sales
