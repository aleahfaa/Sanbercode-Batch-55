<h2>Soal 4: Comparison Operator dengan fungsi</h2>

Buatlah sebuah fungsi yang menerima satu argument bertipe data numeric dan menghasilkan sebuah return sebagai berikut :
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan ganjil
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 2 sampai 5 (2 dan 5 termasuk)
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 6 sampai 20 (6 dan 20 termasuk)
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan lebih besari dari 20

```python
# Buat fungsi
def bil_aneh(angka):
  if angka%2!=0:
    return "Aneh"
  elif angka%2==0 and 2<=angka<=5:
    return "Tidak Aneh"
  elif angka%2==0 and 6<=angka<=20:
    return "Aneh"
  elif angka%2==0 and angka>20:
    return "Tidak Aneh"

print(bil_aneh(22))
```

<h2>Soal 5: While Loop dan For Loop</h2>

- Apa perbedaan while loop dan for loop?
- Berikan contoh sederhana cara menggunakan while loop dan for loop

Perbedaan utama antara loop `for` dan loop `while` adalah bagaimana mereka menangani inisialisasi dan pemeriksaan kondisi. Loop `for` memungkinkan Anda untuk menginisialisasi, memeriksa kondisi, dan melakukan iterasi di dalam header loop, sementara loop `while` membutuhkan Anda untuk menginisialisasi dan memeriksa kondisi di luar loop, dan kemudian melakukan iterasi di dalam badan loop.

```python
#contoh loop for
for i in range(5):
    print(i)

#contoh loop while
i = 0
while i < 5:
    print(i)
    i += 1
```

Pada contoh loop `for`, loop melakukan iterasi dari 0 hingga 4 (inklusif), mencetak setiap angka. Inisialisasi, pemeriksaan kondisi, dan pernyataan iterasi semuanya diatur di dalam header loop.<br>

Pada contoh loop `while`, loop menginisialisasi variabel `i` menjadi 0, dan kemudian terus melakukan iterasi selama `i` kurang dari 5. Inisialisasi dan pemeriksaan kondisi dilakukan di luar loop, dan pernyataan iterasi dieksekusi di dalam badan loop.

<h2>Soal 6: While Loop</h2>
    
Buatlah sebuah code while loop sebagai berikut :
- Buatlah suatu fungsi yang menerima satu input nilai numerik
- Dalam fungsi, Buat sebuah while-loop dengan melakukan comparasi terhadap inputan tersebut, dimana looping/iterasi terus berjalan apabila nilai inputan tidak sama dengan 0
- Dalam setiap iterasi update nilai variable inputan itu, jika input lebih dari 0, maka kurangi variabel tersebut sebesar 1, selain itu tambahkan 1
- print nilai variable input dalam setiap awal iterasi

```python
a = 10

def fungsi_while(angka):
  while angka != 0:
    print(angka)
    if angka>0:
      angka -= 1
    else:
      angka += 1

fungsi_while(a)
```

<h2>Soal 7: For Loop</h2>

Lengkapi kode untuk menghasilkan suatu output yang diharapkan:





*   Cari siswa mana saja yang memiliki nilai lebih dari sama dengan 80. Masukkan kedalam sebuah list. print hasilnya






```python
data_siswa = [
    {
        "nama":"Budi",
        "nilai": 90
    },
    {
        "nama":"Nina",
        "nilai": 78
    },
    {
        "nama":"Rudi",
        "nilai": 91
    },
    {
        "nama":"Olivia",
        "nilai": 76
    },
    {
        "nama":"Leo",
        "nilai": 80
    },
    {
        "nama":"Liam",
        "nilai": 67
    },
    {
        "nama":"Sheila",
        "nilai": 76
    }
]
```

```python
siswa_berprestasi = []

for siswa in data_siswa:
  if siswa["nilai"]>=80:
    siswa_berprestasi.append(siswa["nama"])

print(siswa_berprestasi)
```

['Budi', 'Rudi', 'Leo']

<h2>Soal 8: For Loop Enumerate</h2>
    
Lengkapi kode untuk menghasilkan suatu output yang di harapkan:

- Buatlah sebuah loop dengan mengiterasi sebuah objek list kemudian di kalikan dengan index dari list tersebut dan print hasilnya

```python
obj_list= [1, 16, 11, 10, 5]

# lengkapi code di bawah
for i in range(len(obj_list)):
  hasil=i*obj_list[i]

print(hasil)
```

```python
obj_list= [1, 16, 11, 10, 5]

# lengkapi code di bawah
for index, item in enumerate(obj_list):
  hasil=index*item

print(hasil)
```
