# Data Merging

<h2>Soal 1: Penggabungan Data</h2>

- Sebutkan perbedaan antara full outer join, inner join, dan left join

Perbedaan utama antara full outer join, inner join, dan left join adalah bagaimana mereka menangani pencocokan baris dari dua tabel.

1. **Inner Join**: Jenis join ini mengembalikan hanya baris yang memiliki nilai yang cocok di kedua tabel. Ini juga dikenal sebagai irisan dari kedua tabel.

2. **Left Join**: Jenis join ini mengembalikan semua baris dari tabel kiri dan baris yang cocok dari tabel kanan. Jika tidak ada baris yang cocok di tabel kanan, kolom dari tabel kanan diisi dengan nilai NULL.

3. **Full Outer Join**: Jenis join ini mengembalikan semua baris dari kedua tabel. Jika tidak ada baris yang cocok di salah satu tabel, kolom dari tabel yang tidak ada diisi dengan nilai NULL.

Secara ringkas, inner join mengembalikan hanya baris yang cocok, left join mengembalikan semua baris dari tabel kiri dan baris yang cocok dari tabel kanan, dan full outer join mengembalikan semua baris dari kedua tabel.

<h2>Soal 3: Penggabungan Data</h2>

Lakukan jenis penggabungan data full outer join, inner join, left outer join dan right outer join pada kedua data berikut:

```python
import pandas as pd

data1 = {
    'id' : [1, 2, 3, 4],
    'Feature1' : ['a1', 'a2', 'a3', 'a4'],
    'Feature2' : ['b1', 'b2', 'b3', 'b4']
}

data2 = {
    'id' : [3, 4, 5 ,6, 7],
    'Feature3' : ['c1', 'c2', 'c3', 'c4', 'c5']
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
#join data
df_outer_join = df1.merge(df2, how='outer', on='id')
print(df_outer_join)

df_inner_join = df1.merge(df2, how='inner', on='id')
print(df_inner_join)

df_left_join = df1.merge(df2, how='left', on='id')
print(df_left_join)

df_right_join = df1.merge(df2, how='right', on='id')
print(df_right_join)
```

<h2>Soal 5: Process Grouping</h2>

Jelaskan prosess yang terjadi dalam grouping data menggungkan method <b>groupby()</b> dari pandas.

jawab:

Metode `groupby()` dalam Pandas digunakan untuk mengelompokkan data berdasarkan kolom atau indeks tertentu. Ini mengembalikan objek `DataFrameGroupBy`, yang merupakan kamus di mana kunci-kunci adalah kelompok unik di mana catatan dibagi, dan nilainya adalah objek groupby. Fungsi `groupby()` membagi data menjadi kategori atau kelompok yang berbeda dan menawarkan fleksibilitas untuk menganalisis data dengan menerapkan berbagai metode pada objek groupby.

Berikut adalah penjelasan langkah demi langkah dari proses tersebut:

1. **Membagi data menjadi kelompok**: Metode `groupby()` membagi data menjadi kelompok-kelompok berbeda berdasarkan kolom atau indeks yang ditentukan. Misalnya, jika Anda memiliki DataFrame `df` dan ingin mengelompokkannya berdasarkan kolom 'kategori', Anda dapat menggunakan `df.groupby('kategori')`.

2. **Menerapkan fungsi ke setiap kelompok**: Setelah membagi data menjadi kelompok-kelompok, Anda dapat menerapkan fungsi ke setiap kelompok. Di sinilah fungsi `agg()` berguna. Misalnya, jika Anda ingin menghitung jumlah kolom 'harga' untuk setiap kelompok, Anda dapat menggunakan `df.groupby('kategori').agg(sum)`.

3. **Menggabungkan hasil**: Terakhir, Anda dapat menggabungkan hasil dari aplikasi fungsi untuk mendapatkan output yang diinginkan. Misalnya, jika Anda ingin mendapatkan DataFrame dengan data yang dikelompokkan, Anda dapat menggunakan `df.groupby('kategori').agg(sum).reset_index(name='total_harga')`.

Metode `groupby()` adalah alat yang powerful dalam Pandas yang memungkinkan Anda melakukan berbagai operasi pada data Anda, seperti agregasi, transformasi, dan filtrasi, untuk mendapatkan wawasan dan memahami data Anda lebih baik.

Group data diamond berdasarkan cut (kualitas), lakukan aggregasi dengan fungsi rata2.

<h2>Soal 7: Grouping Some Column base on Multiple Criteria</h2>
    
Lakukan groupby pada data berdasarkan kriteria cut (kualitas) dan color (warna) secara berurutan, dengan aggregasi pada column carat, table, dan price saja. dengan fungsi agregasi rata-rata.

```python
#code here
grouped_df=df.groupby(['cut', 'color'])
aggregated_df=grouped_df[['carat', 'table', 'price']].mean()
print(aggregated_df.head())
```

<h2>Soal 8: Group By Multiple Aggregate Base on Multiple Criteria</h2>
    
Lakukan groupby pada data berdasarkan kriteria cut(kualitas) dan color (warna) secara berurutan kemudian aggregasi objek group by nya dengan fungsi rata2 dan median.

```python
#code here
grouped_df=df.groupby(['cut', 'color'])
aggregated_df=grouped_df.agg({'carat':['mean', 'median'], 'depth':['mean', 'median'], 'table':['mean', 'median'], 'price':['mean', 'median'], 'x':['mean', 'median'], 'y':['mean', 'median'], 'z':['mean', 'median']})
print(aggregated_df)
```

<h2>Soal 9: Group By With Different Aggregation for specific Column</h2>
    
Lakukan groupby pada data berdasarkan criteria cut(kualitas), kemudian aggreagate dengan fungsi max pada column price dan fungsi min pada column carat.

```python
#code here
grouped_df=df.groupby('cut').agg({'price':'max', 'carat':'min'})
print(grouped_df)
```

<h2>Soal 10: Group By With Custom Function</h2>
    
Lakukan groupby pada data berdasarkan criteria cut(kualitas), kemudian lakukan aggregasi dengan fungsi standar deviasi pada column carat dan fungsi rentang nilai antara max dan min pada column price.

```python
#code here
grouped_df=df.groupby('cut').agg({'carat':'std', 'price':lambda x:x.max()-x.min()})
print(grouped_df)
```
