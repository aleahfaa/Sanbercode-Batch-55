<h2>Soal 2: Data Restructuring</h2>

Jelaskan itu pivoting dan melting data

<h3><b>Pivoting</b></h3>

Pivoting adalah operasi transformasi data yang dilakukan menggunakan fungsi `pivot()` dalam perpustakaan Pandas. Pivoting digunakan untuk mengubah bentuk data dari format yang lebar menjadi format yang sempit, di mana baris menjadi kolom dan kolom menjadi baris. Transformasi ini sangat berguna ketika Anda ingin membandingkan berbagai variabel di seluruh entitas dengan lebih efektif.

Fungsi `pivot()` dalam Pandas memungkinkan Anda untuk menentukan indeks, kolom, dan nilai yang akan digunakan dalam DataFrame baru. Parameter indeks menetapkan struktur baris dari DataFrame baru, parameter kolom mengidentifikasi nilai unik mana yang akan berfungsi sebagai kolom baru, dan parameter nilai memastikan bahwa data yang tepat masuk ke dalam struktur baru.

Berikut adalah contoh penggunaan fungsi `pivot()`:

```python
import pandas as pd

df=pd.DataFrame({'A': ['John', 'Boby', 'Mina'], 'B': ['Masters', 'Graduate', 'Graduate'], 'C': [27, 23, 21]})

pivoted_df=df.pivot('A', 'B', 'C')

print(pivoted_df)
```

Ini akan menghasilkan DataFrame yang dipivot di mana 'A' menjadi indeks, 'B' menjadi kolom, dan nilai-nilai di kolom 'C' mengisi sel-sel tabel pivot yang dihasilkan. Fungsi ini mengasumsikan bahwa setiap kombinasi 'A' dan 'B' memiliki nilai yang sesuai yang unik di 'C'.

Pivoting adalah keterampilan penting untuk manipulasi, transformasi, dan analisis data dalam Python. Ini membantu menjaga integritas data dengan menghindari duplikasi yang tidak perlu, meningkatkan analisis data dengan menyusun ulang kumpulan data, dan meningkatkan visualisasi data dengan mengatur data dengan cara yang dapat dengan mudah diinterpretasikan dan memberikan wawasan yang jelas tentang pola dan tren yang mendasarinya.

<h3><b>Melting Data</b></h3>

Melting data merujuk pada proses mengubah bentuk DataFrame dari format yang lebar menjadi format yang panjang menggunakan fungsi `melt()` dalam perpustakaan Pandas. Fungsi `melt()` digunakan untuk mentransformasikan DataFrame di mana satu atau lebih kolom adalah variabel identifikasi (id_vars), sementara semua kolom lainnya, yang dianggap sebagai variabel yang diukur (value_vars), "dipivot" ke sumbu baris, meninggalkan hanya dua kolom non-identifikasi, 'variable' dan 'value'.

Tujuan utama dari melting data adalah untuk mengorganisir data dalam format yang lebih ramah komputer yang lebih mudah untuk dianalisis dan divisualisasikan. Transformasi ini sangat berguna ketika Anda memiliki DataFrame dengan format yang lebar dengan beberapa kolom dan ingin mengonversinya menjadi format yang panjang dengan satu baris untuk setiap pengamatan.

Berikut adalah contoh penggunaan fungsi `melt()`:

```python
import pandas as pd

df = pd.DataFrame({'A': ['John', 'Boby', 'Mina'], 'B': ['Masters', 'Graduate', 'Graduate'], 'C': [27, 23, 21]})

melted_df = pd.melt(df, id_vars=['A'], value_vars=['B'])

print(melted_df)
```

Ini akan menghasilkan DataFrame yang melted di mana 'A' menjadi indeks, 'B' menjadi kolom, dan nilai-nilai di kolom 'C' mengisi sel-sel tabel pivot yang dihasilkan. Fungsi ini mengasumsikan bahwa setiap kombinasi 'A' dan 'B' memiliki nilai yang sesuai yang unik di 'C'.

Melting data adalah keterampilan penting untuk manipulasi, transformasi, dan analisis data dalam Python. Ini membantu menjaga integritas data dengan menghindari duplikasi yang tidak perlu, meningkatkan analisis data dengan menyusun ulang kumpulan data, dan meningkatkan visualisasi data dengan mengatur data dengan cara yang dapat dengan mudah diinterpretasikan dan memberikan wawasan yang jelas tentang pola dan tren yang mendasarinya.

<h2>Soal 4: Reshaping Data 2</h2>

````python
import pandas as pd

data = {
    'Hari' : ['Sabtu', 'Minggu'],
    'Kebun Binatang' : [271, 399],
    'Dufan' : [501, 700],
    'Disney' : [1000, 1001],
    'Bali' : [900, 803]
}
df = pd.DataFrame(data).set_index('Hari')
df
````

diatas adalah data tentang jumlah pengunjung di berbagai tempat hiburan, ubahlah struktur data tersebut sehingga sesuai dengan hasil yang diharapkan.

````python
#code here
melted_df=df.reset_index().melt(id_vars=['Hari'], var_name='Tempat Hiburan', value_name='Pengunjung')
print(melted_df)
````
